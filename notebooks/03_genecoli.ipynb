# Whole-Cell Modeling with genEcoli

This notebook demonstrates how to load and interact with the [genEcoli](https://github.com/vivarium-collective/genEcoli) whole-cell *E. coli* model through the Vivarium interface.

genEcoli wraps 54 biological processes from vEcoli into a process-bigraph Composite, modeling:
- Metabolism (FBA)
- Transcription and translation
- DNA replication
- Cell division
- And more

## 1. Load the genEcoli model

genEcoli provides `load_ecoli_composite()` which loads a pre-built whole-cell model from a saved document. We wrap it in a Vivarium for inspection and simulation.

````python
from vivarium import Vivarium
from genEcoli import load_ecoli_composite
from genEcoli.types import ECOLI_TYPES

def register_ecoli(core):
    """Register E. coli custom types into the core."""
    core.register_types(ECOLI_TYPES)
    return core

# Load the pre-built E. coli composite
# Note: This requires a pre-generated document (see genEcoli documentation)
ecoli_composite = load_ecoli_composite()

# Wrap in Vivarium for interactive use
v = Vivarium(
    document={
        'schema': ecoli_composite.serialize_schema(),
        'state': ecoli_composite.serialize_state(),
    },
    register=register_ecoli,
    emitter='all',
)
print(v)
````

## 2. Inspect the model

Explore the structure and components of the whole-cell model.

````python
# List the top-level state keys
print("Top-level state keys:")
for key in sorted(v.state.keys()):
    print(f"  {key}")
````

````python
# View the registered processes
v.processes(as_dataframe=True)
````

````python
# Visualize the model structure
v.diagram()
````

## 3. Run and analyze

Run a short simulation and plot the results.

````python
# Run a short simulation (10 seconds of biological time)
v.run(10)

# Get timeseries as a DataFrame
df = v.timeseries(as_dataframe=True)
print(f"Timeseries shape: {df.shape}")
df.head()
````

````python
# Plot selected timeseries
fig = v.plot()
fig
````
